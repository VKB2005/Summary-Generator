```python
import pandas as pd # import pandas to help us in data preparation
from datasets import load_dataset
from transformers import T5ForConditionalGeneration, T5Tokenizer, BartForConditionalGeneration, BartTokenizer

max_input_length = 512
max_target_length = 150

def preprocess_function(examples):
    # Assuming your CSV has columns named 'Text' and 'Summary'
    inputs = examples["input"]  # Changed to 'Text'
    targets = examples["output"] # Changed to 'Summary'

    # Convert to strings and replace NaN with an empty string
    inputs = [str(x) if pd.notna(x) else '' for x in inputs]
    targets = [str(x) if pd.notna(x) else '' for x in targets]

    model_inputs = tokenizer(inputs, max_length=max_input_length, padding="max_length", truncation=True)
    # Check if labels need to be a list of lists or just a list
    labels = tokenizer(targets, max_length=max_target_length, padding="max_length", truncation=True).input_ids
    model_inputs["labels"] = labels

    return model_inputs

# Tokenize dataset
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
from transformers import Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=4,   # Adjust based on memory
    gradient_accumulation_steps=2,   # Helps simulate larger batch size
    fp16=True,                       # Mixed precision training
    num_train_epochs=2,              # Reduce for initial testing
    save_steps=500,                  # Adjust to avoid frequent saving
    save_total_limit=2,              # Limit number of checkpoints to save memory
    evaluation_strategy="steps",     # Adjust based on needs
    eval_steps=500                   # Frequency of evaluation
)
```

```python
!pip install nltk

import nltk
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize
from nltk.tokenize.treebank import TreebankWordDetokenizer
# Assume you already have preprocessed tokens of novel and actual summary
# novel_tokens = input  # Replace with your tokenized novel text
actual_summary_tokens = "Evelyn discovers a mysterious journal in her grandmother’s attic, its pages worn and yellowed with age . each entry chronicled the life of a young woman named Clara Hawthorne, a name Evelyn had never heard before . “I saw her again today, the woman in the mirror. She is not me, yet she moves as I do, her eyes pleading as though trapped in the glass. I dare not tell anyone, for who would believe"

# Generate summary from the tokenized novel text
predicted_summary_text = summary

# Tokenize the predicted summary for evaluation
predicted_summary_tokens = predicted_summary_text.split()  # Split the text into tokens

# Initialize the ROUGE scorer
from rouge_score import rouge_scorer
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)

# Function to detokenize the tokens for ROUGE evaluation
def decode_tokens(tokens):
    """
    Detokenizes a list of tokens back into a string.

    Args:
        tokens (list): A list of tokens.

    Returns:
        str: The detokenized string.
    """
    detokenizer = TreebankWordDetokenizer()
    return detokenizer.detokenize(tokens)
# Convert token lists back to strings for ROUGE evaluation
actual_summary_for_rouge = decode_tokens(word_tokenize(actual_summary_tokens)) # Assuming actual_summary_tokens is a string
predicted_summary_for_rouge = decode_tokens(predicted_summary_tokens)

# Calculate ROUGE scores
rouge_scores = scorer.score(actual_summary_for_rouge, predicted_summary_for_rouge)

# Print the ROUGE scores
print("\nROUGE Scores:")
print("ROUGE-1:", rouge_scores["rouge1"])
print("ROUGE-2:", rouge_scores["rouge2"])
print("ROUGE-L:", rouge_scores["rougeL"])

# Calculate BLEU score
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
smoothing = SmoothingFunction().method1
bleu_score = sentence_bleu(
    [word_tokenize(actual_summary_tokens)], predicted_summary_tokens,  # Assuming actual_summary_tokens is a string
    weights=(1, 0, 0, 0),  # Unigram BLEU score
    smoothing_function=smoothing
)
print("\nBLEU Score:", bleu_score)

# Generate the evaluation report
report = f"""
Evaluation Report:
------------------
ROUGE-1: {rouge_scores['rouge1']}
ROUGE-2: {rouge_scores['rouge2']}
ROUGE-L: {rouge_scores['rougeL']}
BLEU Score: {bleu_score}
"""

print("\n" + report)
```
